```python
#Import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from matplotlib.pyplot import figure

%matplotlib inline

matplotlib.rcParams['figure.figsize']=(12,8)  #adjust configuration of plot we will create

#lets import the data 
df= pd.read_csv(r'C:\Users\Fares\Desktop\movies.csv')

```

```python
#let's look at the data
df
```

```python
#checking NUll values , the percentage by column
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print (col, pct_missing)
```

```python
# lets replace every nan with 0 
df = df.fillna(0)
```

```python
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print (col, pct_missing)
    
```

```python
#looking at data types of columns
df.dtypes
```

```python
df.budget
```

```python
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
df['votes']=df['votes'].astype('int64')
df['runtime']=df['runtime'].astype('int64')
```

```python
#looking at the changes, we eliminated every .0 by converting the data type to integer
df
```

```python
#lets sort values by gross
df.sort_values(by='gross',ascending=False)
```

```python
#scatter plot with budget VS gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
#plot using seaborn , Budget vs Gross
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
#lets see coorelations with only numeric numbers #using Pearson which is the default method
df.select_dtypes(include=['number']).corr()
```

```python
#we are looking at a good correlation 0.75 between Budget and Gross, 
#we also have a good coorelation between Gross and Votes
```

```python
#lets run it with another method: Spearman 
```

```python
df.select_dtypes(include=['number']).corr(method='spearman')
```

```python
#Close results with the two methods, we actually got gross and votes have more corellation with this method. 
#Good Correlation between gross and budget , gross and votes
```

```python
#lets see it in a better way with heat map
Coorelation_matrix= df.select_dtypes(include=['number']).corr()
sns.heatmap(Coorelation_matrix, annot=True)
plt.title('Correlation matrix for numeric features')
plt.xlabel('movie feautures')
plt.ylabel('movie feautures')

plt.show()
```

```python
#lets numerize data, we have 'country','name','rating','genre'... a few that we should numerize to see correlation
df_num= df
```
